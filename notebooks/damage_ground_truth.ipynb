```python
import pandas as pd
import geopandas as gpd
import os
import numpy as np
from collections import namedtuple
from IPython.display import display, Markdown, Image
```

```python
# file = 'digitalglobe_crowdsourcing_hurricane_maria_raw_results_20171012'
# file = 'digitalglobe_crowdsourcing_hurricane_maria_20170926.geojson'
file = 'digitalglobe_crowdsourcing_hurricane_irma_20170926.geojson'
df = gpd.read_file(os.path.join(os.pardir, 'data', file))
```

```python
Label = namedtuple('Label', 'link id')
df = df.loc[(df.label == 'Flooded / Blocked Road') | (df.label == 'Blocked Bridge')]
labels = [Label(df.iloc[i].chip_url, df.iloc[i].tag_id)  for i in range(len(df.index))]
df.head()
```

```python
from PIL import Image
import requests
from io import BytesIO

for label in labels:
    response = requests.get(label.link)
    img = Image.open(BytesIO(response.content))
    img.save(os.path.join(os.pardir, 'data/ground_truth', str(label.id)) + '.png')
#     img.show()
```
